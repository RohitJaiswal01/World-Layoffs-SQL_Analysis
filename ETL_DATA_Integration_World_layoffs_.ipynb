```python
import requests
import pandas as pd
from io import StringIO

# URL of the raw CSV file from GitHub
url = 'https://raw.githubusercontent.com/RohitJaiswal01/World-Layoffs-SQL_Analysis/refs/heads/main/layoffs_2023_uncleaned.csv'

# Send GET request to fetch the raw CSV file
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Convert the content of the CSV into a pandas DataFrame
    csv_data = StringIO(response.text)
    df = pd.read_csv(csv_data)
    print(df.head())  # Display first few rows of the DataFrame
else:
    print(f"Failed to fetch CSV. Status code: {response.status_code}")
```

```python
# Saving dataset in a csv format te be processed in MySQL
df.to_csv("layoffs")
df
```
